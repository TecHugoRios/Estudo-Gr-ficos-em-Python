Como analista financeiro, você foi encarregado de examinar as despesas mensais de uma
empresa para verificar variações e identificar possíveis áreas de corte. O conjunto de dados
contém informações sobre as despesas de cada mês nos últimos dois anos. 


Calcule a variação percentual das despesas entre os meses e crie um gráfico de linha para
visualizar as flutuações nas despesas ao longo do tempo

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dados = {'Mês':["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho", "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"],
         'Despesas (R$)':[25000,30000, 32000, 28500, 33000, 31500, 35000, 40000, 38000, 42000, 45000, 50000]}
```

```python
df = pd.DataFrame(dados)
df
```

```python
calcVariacao = [dAtual - dAnterior for dAnterior, dAtual in zip(df['Despesas (R$)'][:-1], df['Despesas (R$)'][1:])]
calcVariacao
```

```python
total = df['Despesas (R$)'].sum()
total
```

```python
plt.figure(figsize=(14,8))
plt.plot(df['Mês'], df['Despesas (R$)'], color='red', linestyle=':',linewidth=3 ,label='Variação', marker=".", markersize=15)

for item, (x,y) in enumerate(zip(df['Mês'][1:], df['Despesas (R$)'][1:])):
    plt.annotate(format(((calcVariacao[item]/total) * 100), ".1f")+ '%', xy=(x, y), xytext=(0, 10), textcoords='offset points', color='black')

plt.ylabel("Despesas (R$)")
plt.xlabel("Percentual de variação a mês anterior")
plt.show()
```
